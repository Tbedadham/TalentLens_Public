# TalentLens – Pre-DS3 Resume Pipeline Plan (Using Kaggle Dataset)

**Dataset:**  
https://www.kaggle.com/datasets/snehaanbhawal/resume-dataset

This public dataset will be used to build and validate the full TalentLens NLP + search pipeline *before* DS3 resumes are available. The goal is to design a modular system where swapping in DS3 data later requires minimal changes.

---

## Phase Goal (Before DS3 Data)

Build an end-to-end, interpretable NLP + information-retrieval pipeline that:
- Ingests resume text (PDF / raw text)
- Cleans, tokenizes, and normalizes unstructured data
- Converts text into feature vectors (TF-IDF)
- Supports query-based candidate ranking (cosine similarity)

All components should be reusable for DS3 resumes later.

---

## Group Task Assignments

### 1. Data Ingestion & Storage

**Purpose:**  
Replicate DS3-style ingestion using the Kaggle dataset.

**Tasks:**
- First: Load resume text from CSV fields
- Then: Extract text from PDFs (pdfplumber). This is because our DS3 data will be PDFs.
- Assign stable candidate IDs
- Store raw text in a standardized folder structure

**Deliverables:**
- Ingestion script/notebook
- Folder of raw resume text files
- README describing data format

---

### 2. Text Cleaning & Tokenization

**Purpose:**  
Convert raw resumes into clean, NLP-ready text.

**Tasks:**
- Lowercasing, whitespace normalization
- Remove boilerplate and scraping artifacts
- Tokenization (NLTK / spaCy / simple rules)
- Optional: section labeling or skill extraction

**Deliverables:**
- Reusable text-cleaning module
- Tokenized and cleaned resume dataset
- Documentation of preprocessing decisions

---

### 3. Feature Extraction & Vectorization

**Purpose:**  
Create numerical representations for matching resumes to queries.

**Tasks:**
- Fit TF-IDF vectorizer on cleaned resumes
- Experiment with resume-only vs combined representations
- Save vectorizer and feature matrix

**Deliverables:**
- Serialized TF-IDF vectorizer
- Candidate feature matrix
- Notebook showing feature construction

---

### 4. Query & Ranking Engine

**Purpose:**  
Implement the core search and ranking logic.

**Tasks:**
- Vectorize recruiter queries
- Compute cosine similarity with resume vectors
- Rank candidates by relevance
- Output ranked results with scores

**Deliverables:**
- Ranking function or module
- Demo notebook (query → ranked candidates)
- Example query outputs

---

### 5. Evaluation & Prototype Interface

**Purpose:**  
Validate relevance and usability.

**Tasks:**
- Define relevance metrics (top-K precision, manual labels)
- Test queries against known resume categories
- Build a simple prototype interface (CLI, notebook, or Streamlit)

**Deliverables:**
- Evaluation report
- Simple query interface
- Notes on failure cases and bias

---

## Mapping to TalentLens Project Scope

| TalentLens Requirement | Kaggle Dataset Work |
|------------------------|---------------------|
| Resume ingestion | CSV/PDF parsing |
| Text parsing & cleaning | Cleaning + tokenization |
| Feature extraction | TF-IDF vectors |
| Search & ranking | Cosine similarity |
| Evaluation | Manual relevance testing |
| Interface | Notebook or Streamlit demo |

---

## Transition to DS3 Data (Later)

When DS3 resumes become available:
- Replace Kaggle resume input with DS3 resume corpus
- Re-run ingestion and preprocessing
- Reuse TF-IDF, ranking, and evaluation logic
- Minimal or no refactoring required

Design rule:
> The pipeline should not care where the resume text comes from.

---

## Key Takeaway

This Kaggle dataset enables the team to build, test, and validate the *entire* TalentLens pipeline early, ensuring a smooth transition once private DS3 data is introduced.
