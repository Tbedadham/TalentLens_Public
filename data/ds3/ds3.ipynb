# Member Resumes

I've upated the member.txt file to include the other columns of the members in the DS3 database along with their resume links. The resumes you guys were supposed to scrape are unchanged, and none were added yet. The SQL command in the cell below has the columns that now are also given to us and are in member.txt.

SELECT full_name, email, graduation_year, major, resume_link, linkedin_link, github_link, other_link
FROM "Members"
WHERE 
  in_talent_pool AND
  resume_link <> '';
